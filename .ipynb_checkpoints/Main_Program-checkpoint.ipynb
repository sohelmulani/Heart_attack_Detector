```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#these are some custom classes mannualy created for data preprocessing,feature_selection,data visualisation and metrics
from pre_processing import Preprocessing
from feature_selection import FeatureSelection
from visualisation import Vis
from metrics import Metrics
prp=Preprocessing()
fs=FeatureSelection()
vs=Vis()
mtr=Metrics()
```

```python
#importing dataset
dataset=pd.read_csv("heart.csv")
dataset.head()
```

```python
#pairplot visualisation of data.this data visualization methode is provided by seaborn library
vs.seabornVis(dataset)
```

```python
#histogram visualisation of data 
vs.histogramVis(dataset)
```

```python
#deviding dataset into dependant(y) and independant(x)
x=dataset.iloc[:,1:13].values
y=dataset.iloc[:,13]
```

```python
""""
features selection
n is the no of features you want for your model
""""
n=int(input("enter how many features do you wants?"))
selected=fs.selectBest(x,y,n)
```

```python
""""
it will print the score of each feature as related to target
""""
fs.printScore()
```

```python
""""
it will create a matrix of selected features
"""
finalX=fs.selectFinalFetures()
finalX.shape
```

```python
""""
scalling data in the range of 0 and 1
"""
scaledX=prp.scalling(finalX)
```

```python
"""
decomposing data into training and testing set
"""
x_train,x_test,y_train,y_test=prp.decomposition(scaledX,y)
```

```python
"""
I have used nearlly all classification methods in this dataset but KNN  is
giving more accuracy so i am using kNN
"""
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

```python
#predicting values for testing set
y_pred=classifier.predict(x_test)
```

```python
"""
confusion matrix
"""
mtr.confusionMatrix(y_pred,y_test)
```

```python
#printing accuracy of values predicted by kNN model
mtr.accuracy(y_pred,y_test)
```
